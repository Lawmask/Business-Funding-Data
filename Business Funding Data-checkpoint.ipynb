```python
df['Effective date'] = df['Effective date'].fillna('Unknown')

df['Financing Type'] = df['Financing Type'].fillna('Not Specified')

df['Financing Type Normalized'] = df['Financing Type Normalized'].fillna('Not Specified')

df['Investors'] = df['Investors'].fillna('Unknown')

df['Investors Count'] = df['Investors Count'].fillna(0)
```

```python
df['Effective date'] = pd.to_datetime(df['Effective date'], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.isnull().sum()
```

```python
df.to_csv("cleaned_business_funding_data.csv", index=False)
```
